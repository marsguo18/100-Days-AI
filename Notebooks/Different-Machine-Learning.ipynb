## Supervised learning 
**Supervised learning** is the machine learning task of getting the information from **labeled** datasets.
### Example:
- You get bunch of photos with informaton which are **dogs** and you train a model to recognize new photos whether they are dogs.

## Unsupervised learning
**Unsupervised learning** is the machine learning task of getting the datasets **without any labels** given.
### Example:
- You get bunch of photos part of them are **dogs** and part are not. But they have **no information** on them. And you train a model to discover dogs.

## Reinforcement learning
**Reinforcement learning** is the machine learning task of taking suitable action to maximize reward in a particular situation.
### Example:
- You get bunch of photos part of them are dogs and part are not. Just like the example above, they have no information on them at first. But the machine will get **positive** feedbacks (reward) once it select the dogs' photos, get **negative** feedbacks (punishment) once it select others. Then the feedbacks will become the information, helping machine to recognize dogs better and better.
